Implement Gradient Descent Algorithm to find the local minima of a function. 
For example, find the local minima of the function y=(x+3)Â² starting from the point x=2.

```python
# Gradient Descent Implementation in Python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return (x + 3)**2

# Define the derivative of the function
def df(x):
    return 2 * (x + 3)

# Gradient Descent with convergence criterion
def gradient_descent(starting_point, learning_rate, tol=1e-6, max_iter=1000):
    x = starting_point
    history = [x]  # To store the values for plotting
    for i in range(max_iter):
        grad = df(x)
        x_new = x - learning_rate * grad
        history.append(x_new)
        if abs(x_new - x) < tol:  # Convergence check
            break
        x = x_new
    return x, history

# Parameters
starting_point = 2
learning_rate = 0.1

# Run gradient descent
min_x, history = gradient_descent(starting_point, learning_rate)
min_y = f(min_x)

print(f"Local minimum occurs at x = {min_x}, y = {min_y}")

# Plotting the function and the gradient descent path
x_vals = np.linspace(-10, 5, 100)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label='y=(x+3)^2')
plt.scatter(history, [f(x) for x in history], color='red', label='Gradient Descent Steps')
plt.plot(history, [f(x) for x in history], color='red', linestyle='--', alpha=0.5)  # Connect steps
plt.title('Gradient Descent to Find Local Minimum')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
